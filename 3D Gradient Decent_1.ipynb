# Caluclation of MSE (Mean Squared Error)
# $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2$$

```python
def mse(y, y_hat):
    mse_calc = 1/7 * sum((y-y_hat)**2)
    return mse_calc

print(' The directly calculated value of MSE is ', mse(y_1, y_hat))
```

```python
def mse_1(y, y_hat):
    mse_calc_1 = (1/y.size) * sum((y - y_hat)**2)
    return mse_calc_1
print(' The  calculated value of MSE is using y.size is ', mse_1(y_1, y_hat))
```

```python
def mse_2(y, y_hat):
    mse_calc_2 = np.average((y - y_hat)**2, axis=0)
    return mse_calc_2
print ('The MSE Calculated using the np.average is ', mse_2(y_1, y_hat) )
```

```python
print('MSE obtained from inbuilt function in sklearn', mean_squared_error(y_1, regr.predict(x_1)))
```

#### My variables are the thetas
#### theta_0 represented by th_0
#### theta_1 represented by th_1
#### total number of thetas will be nr_thetas
#### doing a mesh grid b/n th_0 and th_1
#### if nr_thetas = 5, then th_0 = 5 and th_1 = 5
#### hence the number of points on the mesh grid will be 5 x 5 = 25

```python
nr_thetas = 5
th_0 = np.linspace(start=-1, stop=3, num=nr_thetas)
th_1 = np.linspace(start=-1, stop=3, num=nr_thetas)
plot_t0, plot_t1 = np.meshgrid(th_0, th_1)
```

```python
plt.figure(figsize=[5, 5])
plt.plot(plot_t0, plot_t1, marker ='s', markeredgewidth=10, )
plt.ylim(-1, 3)
plt.xlim(-1, 3)
plt.grid()
plt.show()
```

```python
plot_cost = np.zeros((nr_thetas, nr_thetas)) 
plot_cost
```

```python
plot_cost = np.zeros((nr_thetas, nr_thetas ))

for i in range(nr_thetas):
    for j in range(nr_thetas):
        print(f' At row:{i}, column:{j},the value of plot_t0 is           :',plot_t0[i][j])
        print(f' At row:{i}, column:{j},the value of plot_t1 is           :' ,plot_t1[i][j])
        y_hat = plot_t0[i][j] + plot_t1[i][j]*x_1
        plot_cost[i][j] = mse(y_1, y_hat)
        print(f' At row:{i}, column:{j},the value of plot_cost is         :',plot_cost[i][j], '\n')

print('Shape of plot_t0 is  ', plot_t0.shape, '\n')
print('Shape of plot_t1 is  ', plot_t1.shape, '\n')
print('Shape of plot_cost is', plot_cost.shape, '\n')
print('value of plot_cost', plot_cost.min())
```

```python
fig = plt.figure(figsize=[16, 12])
ax = fig.gca(projection='3d')
ax.set_xlabel('Theta_0 The Intercept', fontsize=20, color ="black")
ax.set_ylabel('Theta_1 The Slope', fontsize=20, color ='green')
ax.set_zlabel('Cost function - The Mean Square Error', fontsize=20, color = 'brown')
ax.plot_surface(plot_t0, plot_t1, plot_cost, cmap=cm.cool, alpha =1  )
ax.scatter(plot_t0, plot_t1, plot_cost, s=100, color='red', alpha=0.9)
ax.azim = -15    
plt.show()
```

```python
nr_thetas = 200
th_0 = np.linspace(start=-1, stop=3, num=nr_thetas)
th_1 = np.linspace(start=-1, stop=3, num=nr_thetas)
plot_t0, plot_t1 = np.meshgrid(th_0, th_1)
```

```python
plot_cost = np.zeros((nr_thetas, nr_thetas ))
for i in range(nr_thetas):
    for j in range(nr_thetas):
        y_hat = plot_t0[i][j] + plot_t1[i][j]*x_1
        plot_cost[i][j] = mse(y_1, y_hat)
```

```python
fig = plt.figure(figsize=[16, 12])
ax = fig.gca(projection='3d')

ax.set_xlabel('Theta_0 The Intercept', fontsize=20, color ="black")
ax.set_ylabel('Theta_1 The Slope', fontsize=20, color ='green')
ax.set_zlabel('Cost function - The Mean Square Error', fontsize=20, color = 'brown')
ax.plot_surface(plot_t0, plot_t1, plot_cost, cmap=cm.cool, alpha = 0.5  )
ax.azim = -10    
plt.show()
```

```python
def grad(x, y, thetas):
    n = y.size
    
    #Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)
    
    return np.array([theta0_slope[0], theta1_slope[0]])
```

```python
multiplier = 0.01
thetas = np.array([2.9, 2.9])


for i in range(1000):
    thetas = thetas - multiplier * grad(x_1, y_1, thetas)
    
# Results
print('Min occurs when theta 0 is:', thetas[0])
print('Min occurs when theta 1 is:', thetas[1])
print('MSE is means cost function is: ', mse(y_1, thetas[0]+thetas[1]*x_1) )
```

```python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

mse_vals = mse(y_1, thetas[0]+thetas[1]*x_1)
plot_vals = thetas.reshape(1,2)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_1, y_1, thetas)
    mse_vals = np.append(arr = mse_vals , values = mse(y_1, thetas[0]+thetas[1]*x_1))
    plot_vals = np.concatenate((plot_vals,thetas.reshape(1,2)), axis = 0)
```

```python
fig = plt.figure(figsize=[16, 12])
ax = fig.gca(projection='3d')
ax.set_xlabel('Theta 0', fontsize=20)
ax.set_ylabel('Theta 1', fontsize=20)
ax.set_zlabel('Cost - MSE', fontsize=20)
ax.azim = -30
ax.plot_surface(plot_t0, plot_t1, plot_cost, cmap=cm.cool, alpha = 0.5  )
ax.scatter(plot_vals[:, 0], plot_vals[:, 1], mse_vals, s=40, color='red')
plt.show()
```
