```python
# making a dataframe from the data
data = pd.DataFrame(data = boston_dataset.data, columns = boston_dataset.feature_names)
data
```

```python
data['PRICE'] = boston_dataset.target # adding target feature to dataframe
```

```python
data
```

```python
plt.hist(data['PRICE'], bins =30, ec = 'black' )
plt.show()
```

```python
plt.figure(figsize= (10,6))
plt.hist(data['PRICE'], bins =50, ec = 'black', color = '#A300FF', alpha = 0.7)# bins 50...setting color by  hexcode
plt.xlabel('Price in 1000\'s')
plt.ylabel('Nr of Houses')
plt.show()
```

```python
import seaborn as sns
sns.distplot(data['PRICE'])
plt.show()
```

```python
# Accessibility to high ways....RAD

plt.figure(figsize= (10,6))
plt.hist(data['RAD'], ec = 'black', color = 'green', alpha = 0.4)
plt.xlabel('Accessibility to Highways')
plt.ylabel('Nr of Houses')
plt.show()
```

```python
# doing futher investigation on the RAD value....
# set my bins to max value of the RAD....that means bins = 24
plt.figure(figsize= (10,6))
plt.hist(data['RAD'], ec = 'black', bins = 24, color = 'green', alpha = 0.5)
plt.xlabel('Accessibility to Highways')
plt.ylabel('Nr of Houses')
plt.show()
```

```python
accessibility = data['RAD'].value_counts()
print(accessibility) 
print(type(accessibility))
```

```python
# plotting a chart a bar chart
plt.figure(figsize= (10,6))
plt.bar(accessibility.index, accessibility )
plt.show()
```

```python
# rugplots
plt.figure(figsize= (10,6))
sns.distplot(data['PRICE'], color = 'red', bins = 50,  norm_hist=False, rug=True, hist = False)
plt.show()
```

```python
# some more visualizations....
# A scatter plot between DIS and NOX
# after doing a corr
nox_dis_corr = round(data['NOX'].corr(data['DIS']),3)
nox_dis_corr
```

```python
plt.figure(figsize = (16, 8))
plt.title (f'Dist from Employment Centres vs Pollution , Corr ({nox_dis_corr})',fontsize = 25, color = 'green'  )
plt.scatter(x=data['DIS'], y =data['NOX'], alpha = 0.5, color = 'red', s= 80)
plt.xlabel( 'Distance from Employment Centres...DIS', fontsize =14, color = 'green' )
plt.ylabel( 'Level of Pollutants Nitric Oxide...NOX', fontsize =14, color = 'green' )
plt.show()
```

```python
# Using seaborn to carry out the plotting...
sns.set()
sns.set_style('dark')
sns.jointplot(x=data['DIS'], y =data['NOX'], color = 'red', joint_kws = {'alpha' : 0.5},)
plt.show()
```

```python
sns.set()
sns.set_style('dark')
plt.figure(figsize = (16, 8))
sns.scatterplot(x=data['DIS'], y =data['NOX'], color = 'red', alpha = 0.5)
plt.show()
```

```python
sns.set()
sns.set_style('dark')
sns.jointplot(x=data['DIS'], y =data['NOX'], kind = 'hex')
plt.show()
```

```python
# RAD vs TAX
sns.set() # to reset any previous styling
sns.set_style('dark')# to set style.
sns.jointplot(x=data['TAX'], y =data['RAD'], color = 'darkred', joint_kws = {'alpha' : 0.5} ) # jointplot gives  scatter plot 
plt.show()
```

```python
# plotting the regressioin line using sns lmplot
sns.lmplot(x ='TAX', y = 'RAD', data = data)
plt.show()
```

```python
sns.set() # to reset any previous styling
sns.set_style('dark')# to set style.
# height  = 7 is optional
sns.jointplot(x=data['RM'], y =data['PRICE'], color = 'indigo', kind = 'scatter', height = 7) # jointplot gives scatter plot 
plt.show()
```

```python
# All defaults regressions with lmplot......
# height = 10 is optional
sns.lmplot(x ='RM', y = 'PRICE', data = data, height =10, aspect = 2)
plt.show()
```

```python
# without the confidence interval that is ci = None
sns.lmplot(x ='RM', y = 'PRICE', data = data, height =10, aspect = 2, ci = None )
plt.show()
```

```python
# plots at one go using seaborn
sns.pairplot(data)
plt.show()
```
