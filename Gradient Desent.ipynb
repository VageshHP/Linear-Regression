# $$f(x)=x^2+x+1$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2+x+1
```

```python
def df(x):
    return 2*x+1
```

```python
plt.figure(figsize=(12,8))
plt.xlabel('X',fontsize=20)
plt.ylabel('df(X)',fontsize=20)
plt.plot(x_1,df(x_1))
plt.show()
```

# GRADIENT DESCENT

```python
new_x=3
prev_x=0
gamma=0.1
x_list=[new_x]
slope_list=[df(new_x)]
for n in range(75):                   # more number of iteration for better minima
    prev_x=new_x
    new_x=prev_x-gamma*df(new_x)
    x_list.append(new_x)
    slope_list.append(df(new_x))
print(new_x)
#print(x_list)
#print(slope_list)
```

```python
plt.figure(figsize=(8,5))
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1))
plt.scatter(x_list, f(np.array(x_list)), color = 'red', s = 100, alpha=0.4)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.plot(x_1, df(x_1))
plt.scatter(x_list, slope_list, color = 'green', s =100, alpha = 0.7)
plt.show()
```

```python
plt.figure(figsize=[20, 5])

# Sub Plot 1
plt.subplot(1, 2, 1)
plt.xlim(-3, 3)
plt.ylim(0, 8)
plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1), color='blue', linewidth=3, alpha=0.8)
values = np.array(x_list)
plt.scatter(x_list, f(values), color='red', s=100, alpha=0.6)

# sub plot2
plt.subplot(1, 2, 2)
plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)
plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

plt.show()
```

```python
def gradient_descent(deriv_func,initial_guess,multiplier=0.1,precision=0.01,max_iter=75):
    new_x=initial_guess
    prev_x=0
    x_list=[new_x]
    slope_list=[df(new_x)]
    for i in range(max_iter):
        prev_x=new_x
        grad=multiplier*deriv_func(new_x)
        new_x=prev_x-grad
        
        step_size=abs(new_x-prev_x)
        x_list.append(new_x)
        slope_list.append(deriv_func(new_x))
        
        if step_size<precision:
            break
    return new_x,x_list,slope_list
```
