```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('cost_revenue_clean.csv')
```

```python
X=data[['production_budget_usd']]
y=data[['worldwide_gross_usd']]
```

```python
plt.figure(figsize=(10,6))                                      # bigger figure size
plt.scatter(X, y, alpha=0.5)                                    # setting the alpha
plt.title('Film Cost vs Global Revenue')                        # Title of the plot
plt.xlabel('Production Budget $')                               # Label on X Axis
plt.ylabel('Worldwide Gross $')                                 # Label on Y axis
plt.ylim(0, 3000000000)                                         # setting the limits of y
plt.xlim(0, 450000000)                                          # setting the limits of X
plt.show()
```

```python
regr = LinearRegression()
regr.fit(X,y)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.2, c = 'blue')
plt.plot(X, regr.predict(X), color='brown', linewidth=2)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget ')
plt.ylabel('Worldwide Gross ')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
print('Intercept:', regr.intercept_)
print('Coef     :', regr.coef_)
print('Accuracy :', regr.score(X,y))
```
